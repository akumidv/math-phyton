```python
## 11-4 Тренировочный тест: Вероятность

# Предположим, что в некоторой популяции до 60 лет доживает 50%, а до 80 лет — 20%. 
# Какова вероятность (от 0 до 1), что случайно выбранный шестидесятилетний представитель популяции доживёт до восьмидесяти? 
# Запишите ответ с точностью до одного знака после десятичной точки.

A = 0.2
B = 0.5


# P(A|B) = P(AB)/P(B), для случа вероятности P(AB) = PA
def pAB(a, b):
    return a/b

print 'Вопрос 1', pAB(A, B)
```

```python
# Среди жителей Австралии группы крови распределены следующим образом:
# P(I) = 0,49,P(II)=0.38, p(III)=0.1, P(IV)=0.03
# Какова вероятность того, что плазму одного случайно выбранного австралийца можно перелить другому 
# случайно выбранному австралийцу
P1 = 0.49
P2 = 0.38
P3 = 0.1
P4 = 0.03
def pOther(a, b, c, d):
    return a * a + a * b + a * c + a * d

## Неправильно. Считаем, что не учитывают таблицу совместимости
print 'Вероятность что перельют перую группу, в другую', pOther(P1, P2, P3, P4)
print 'Вероятность что перельют вторую группу, в другую', pOther(P2, P1, P3, P4)
print 'Вероятность что перельют третью группу, в другую', pOther(P3, P1, P2, P4)
print 'Вероятность что перельют четвертую группу, в другую', pOther(P4, P1, P2, P3)
print 'Вероятность что перельют любую группу, в другую', pOther(P1, P2, P3, P4) + pOther(P2, P1, P3, P4) + pOther(P3, P1, P2, P4) + pOther(P4, P1, P2, P3)

print 'Вероятность, что первой перельют, первую', P1*P1
print 'Вероятность, что второй перельют, вторую', P2*P2
print 'Вероятность, что третьей перельют, третью', P3*P3
print 'Вероятность, что четвертой перельют, четветую', P4*P4
print 'Вероятность что перельют правильной группе, её же', P1*P1 + P2*P2 + P3*P3 + P4*P4
print 'Обратная вероятность что перельют правильной группе, её же', 1 - (P1*P1 + P2*P2 + P3*P3 + P4*P4)
# неправильный ответ
```

```python
## Задание 3 
# В супермаркете 60% яблок из Турции и 40% яблок из Индии. 
# 10% турецких и 15% индийских яблок — червивые. 
# Какова вероятность, что яблоко, купленное в этом магазине, окажется червивым? 
PT = 0.6
PI = 0.4
PTch = 0.1
PIch = 0.15
print 'Вероятность, что купленное яблоко турецкое и червивое', PT * PTch
print 'Вероятность, что купленное яблоко индийское и червивое', PI * PIch
print 'Вероятность, что любое червивое', PT * PTch + PI * PIch
```

```python
## Задача 4
# Представьте, что вы купили яблоко в магазине из предыдущей задачи, и оно оказалось червивым. 
# Какова вероятность того, что оно из Турции, а не из Индии?
print 'Вероятность червивого яблока из турции, вар.1', PT * (PT * PTch + PI * PIch) # Неправильный ответ
print 'Вероятность червивого яблока из турции, вар.2', PT * PTch  * (1 - PI * PIch) # Неправильный ответ
print 'Вероятность червивого яблока из турции, вар.3', (PT * PTch)/(PT * PTch + PI * PIch)
```
